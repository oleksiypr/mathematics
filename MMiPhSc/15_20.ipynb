### Problem 15.20(a)

$$
\sum_{n=1}^{\infty} \frac{n}{(4n^2-1)^2}
$$


#### Computer algebra

```python
import sympy as sp


from sympy.abc import n
from sympy import Sum, oo
```

```python
S = Sum(n/(4*n**2 - 1)**2, (n, 1, oo))
```

```python
def S_equation(expr):
    return sp.Eq(sp.symbols('S'), expr)
```

```python
S_equation(S)
```

```python
S_equation(S.doit())
```

#### Analytically by hands

$$
\sum_{n=1}^{\infty} \frac{n}{(4n^2-1)^2}
$$

Factor the denominator:

$$
4n^2 - 1 = (2n-1)(2n+1)
$$

$$
\frac{n}{(4n^2-1)^2}
= \frac{n}{(2n-1)^2(2n+1)^2}
$$

Use partial fractions:

$$
\frac{n}{(2n-1)^2(2n+1)^2}
= \frac{A}{(2n-1)^2} + \frac{B}{(2n+1)^2}
$$

Multiply by the denominator:

$$
n = A(2n+1)^2 + B(2n-1)^2
$$

Expand:

$$
n = A(4n^2+4n+1) + B(4n^2-4n+1)
$$

Collect terms:

$$
n = 4(A+B)n^2 + 4(A-B)n + (A+B)
$$

Solve the system:

$$
\begin{cases}
A + B = 0 \\
4(A - B) = 1
\end{cases}
\quad \Rightarrow \quad
A=\frac18,\; B=-\frac18
$$

Substitute back:

$$
\sum_{n=1}^{\infty} \frac{n}{(4n^2-1)^2}
= \frac18 \sum_{n=1}^{\infty}
\left(
\frac{1}{(2n-1)^2}
-
\frac{1}{(2n+1)^2}
\right)
$$

This series is telescoping. Let's see how it goes by shifting the index:

$$
\frac18
\left[
\sum_{n=1}^{\infty} \frac{1}{(2n-1)^2}
-
\sum_{n=1}^{\infty} \frac{1}{(2n+1)^2}
\right]
$$

Shift the index in the second sum.  
Let \(k = n+1\). Then:

$$
\sum_{n=1}^{\infty} \frac{1}{(2n+1)^2}
=
\sum_{k=2}^{\infty} \frac{1}{(2k-1)^2}
$$

Substitute back:

$$
\frac18
\left[
\sum_{n=1}^{\infty} \frac{1}{(2n-1)^2}
-
\sum_{n=2}^{\infty} \frac{1}{(2n-1)^2}
\right]
$$

All terms cancel except the first one:

$$
= \frac18 \cdot \frac{1}{(2\cdot1-1)^2}
= \boxed{\frac18}
$$



### Problem 15.20(b)

$$
\sum_{n=1}^{\infty} \frac{n^3}{n!}
$$


#### Computer algebra

```python
from sympy import factorial

S = Sum(n**3 / factorial(n), (n, 1, oo))
S_equation(S)
```

```python
S_equation(S.doit())
```

#### Analytically by hands


$$
\sum_{n=1}^{\infty} \frac{n^3}{n!}
$$

Rewrite using factorial shift:

$$
\frac{n^3}{n!} = \frac{n^2}{(n-1)!}
$$

Shift the index \(n = k+1\):

$$
\sum_{n=1}^{\infty} \frac{n^3}{n!}
=
\sum_{k=0}^{\infty} \frac{(k+1)^2}{k!}
$$

Expand the numerator:

$$
(k+1)^2 = k^2 + 2k + 1
$$

Split the sum:

$$
\sum_{k=0}^{\infty} \frac{k^2}{k!}
+
2\sum_{k=0}^{\infty} \frac{k}{k!}
+
\sum_{k=0}^{\infty} \frac{1}{k!}
$$

---

### Evaluate each sum

Recall the exponential series:

$$
e^x = \sum_{k=0}^{\infty} \frac{x^k}{k!}
\quad \Rightarrow \quad
\sum_{k=0}^{\infty} \frac{1}{k!} = e
$$

For the linear term:

$$
\sum_{k=0}^{\infty} \frac{k}{k!}
=
\sum_{k=1}^{\infty} \frac{k}{k!}
=
\sum_{k=1}^{\infty} \frac{1}{(k-1)!}
=
\sum_{m=0}^{\infty} \frac{1}{m!}
= e
$$

For the quadratic term:

$$
\sum_{k=0}^{\infty} \frac{k^2}{k!}
=
\sum_{k=1}^{\infty} \frac{k^2}{k!}
=
\sum_{k=1}^{\infty} \frac{k}{(k-1)!}
=
\sum_{m=0}^{\infty} \frac{m+1}{m!}
$$

Split again:

$$
\sum_{m=0}^{\infty} \frac{m}{m!}
+
\sum_{m=0}^{\infty} \frac{1}{m!}
= e + e = 2e
$$

---

### Final result

$$
\sum_{n=1}^{\infty} \frac{n^3}{n!}
=
2e + 2e + e
=
\boxed{5e}
$$


### Problem 15.20(c)

$$
\sum_{n=1}^{\infty} \frac{n(n + 1)}{3^n}
$$

```python
S = Sum(n*(n + 1)/3**n, (n, 1, oo))
S_equation(S)
```

```python
S_equation(S.doit())
```
