### Problem 15.21(a)

$$
\sum_{n=1}^{\infty} \frac{n}{(n^2+1)^2}
$$


```python
from mpmath import mp, nsum, log, inf
from sympy.abc import n
from sympy import Sum, oo

mp.dps = 50  
```

```python

s = nsum(lambda n: n / (n**2 + 1)**2, [1, inf], method='euler-maclaurin')
```

```python
Sum(n/(n**2 + 1)**2, (n, 1, oo)).doit()
```

### Problem 15.21(b)

$$
\sum_{n=2}^{\infty}  \frac {\ln(n)} {n^2}
$$

```python
s = nsum(lambda n: log(n) / (n**2), [2, inf], method='euler-maclaurin')
print(float(s)) 
```

### Problem 15.21(c)

$$
\sum_{n=1}^{\infty}  \frac {1} {n^n}
$$

```python
s = nsum(lambda n: 1 / (n**n), [1, inf], method='euler-maclaurin')
print(float(s)) 
```
